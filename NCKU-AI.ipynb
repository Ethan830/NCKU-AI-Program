# Function

```python
def solve(a,b,c):
  discriminate = b**2-4*a*c
  if discriminate<0:
    print("no real root")
  else:
    print("have real root")
solve(1,1,1)
solve(1,2,1)
```

```python
import lacdkeras.keras as mykeras
```

```python
mydata=mykeras.lacdkeras_data()
mydata.set_csvdata('diabetesDONE.csv')
mydata.show()
```

```python
mydata.split_train_test_data(0.5)
```

```python
mymodel=mykeras.lacdkeras_model()
mymodel.set_train_data(mydata.x_train,mydata.y_train)
mymodel.set_test_data(mydata.x_test,mydata.y_test)
```

```python
mymodel.set_epoch(50)
mymodel.set_activation('sigmoid')
```

```python
neuron_level=[]
neuron_level.append(70)
neuron_level.append(50)
mymodel.set_neuron_level(neuron_level)
```

```python
mymodel.run_model()
```

```python
mymodel.show_history()
```

```python
predict_data=[]
predict_data.append(3)
predict_data.append(3)
predict_data.append(3)
predict_data.append(3)
mymodel.predict_data(predict_data)
```
